## 1. 인덱싱 (Indexing)

인덱싱은 문자열, 리스트, 튜플 등의 시퀀스 자료형에서 **하나의 요소**를 가져오는 방법입니다.  
파이썬은 **0부터 시작**하는 인덱스를 사용합니다.

```python
# 문자열 인덱싱
language = "Python Programming"

print("첫 글자:", language[0])
print("다섯 번째 글자:", language[4])
print("마지막 글자:", language[-1])
print("뒤에서 세 번째 글자:", language[-3])
```

```python
# 리스트 인덱싱
colors = ["red", "green", "blue", "yellow"]

print("첫 번째 색상:", colors[0])
print("마지막 색상:", colors[-1])
print("두 번째 색상을 대문자로 출력:", colors[1].upper())
```

## 2. 슬라이싱 (Slicing)

슬라이싱은 시퀀스 자료형에서 **범위를 지정해 부분 추출**하는 방법입니다.  
형식: `시작:끝:간격` (끝은 포함되지 않음)

- `[:3]`: 처음부터 인덱스 2까지
- `[1:4]`: 인덱스 1부터 3까지
- `[::-1]`: 역순


```python
# 문자열 슬라이싱
sentence = "Data Analysis with Python"

print("처음 4글자:", sentence[:4])
print("5~12번째 글자:", sentence[5:13])
print("Python만 추출:", sentence[-6:])
print("짝수 인덱스 글자:", sentence[::2])
print("문자열을 역순으로:", sentence[::-1])
```

```python
# 리스트 슬라이싱
numbers = [10, 20, 30, 40, 50, 60, 70]

print("처음 3개:", numbers[:3])
print("중간 3개:", numbers[2:5])
print("3칸씩 건너뛰기:", numbers[::3])
print("역순 출력:", numbers[::-1])
```

## 3. 리스트 (List)

- 순서 있음
- 변경 가능 (mutable)
- 다양한 자료형 저장 가능

```python
fruits = ["apple", "banana", "cherry"]

# 추가, 삭제, 삽입
fruits.append("orange")          # 마지막에 추가
fruits.insert(1, "kiwi")         # 중간 삽입
fruits.remove("banana")          # 특정 값 제거
popped = fruits.pop()            # 마지막 값 꺼내기

print("과일 리스트:", fruits)
print("꺼낸 과일:", popped)

# 정렬 및 기타
fruits.sort()                    # 오름차순 정렬
print("정렬된 리스트:", fruits)

fruits.reverse()                 # 역순 정렬
print("역순 리스트:", fruits)

print("과일 수:", len(fruits))
print("cherry의 위치:", fruits.index("cherry"))
```

## 4. 튜플 (Tuple)

- 순서 있음
- 변경 불가 (immutable)
- 리스트와 거의 같지만, 읽기 전용일 때 사용

```python
scores = (88, 92, 75, 88, 100)

print("전체 점수:", scores)
print("점수 개수:", len(scores))
print("88의 개수:", scores.count(88))
print("75의 위치:", scores.index(75))

# 튜플 언팩
a, b, c, d, e = scores
print("첫 번째 점수:", a)
```

## 5. 딕셔너리 (Dictionary)

- 키(key)와 값(value)의 쌍
- 순서 있음 (Python 3.7+)
- 키는 중복 불가

```python
person = {
    "name": "Jane",
    "age": 28,
    "city": "Busan"
}

# 값 읽기, 수정, 추가
print("이름:", person["name"])
person["age"] = 29
person["job"] = "Designer"

# 주요 메서드
print("키 목록:", list(person.keys()))
print("값 목록:", list(person.values()))
print("키-값 쌍:", list(person.items()))

# 안전하게 가져오기
print("국가 정보:", person.get("country", "정보 없음"))

# 삭제
del person["city"]
print("삭제 후:", person)
```

## 6. 리스트 vs 튜플 vs 딕셔너리 비교표

| 항목        | 리스트 (List)       | 튜플 (Tuple)        | 딕셔너리 (Dictionary)     |
|-------------|----------------------|----------------------|----------------------------|
| 구조        | 순서 있는 요소 집합   | 순서 있는 요소 집합   | 키:값 쌍의 집합            |
| 가변성      | ✅ 변경 가능         | ❌ 변경 불가         | ✅ 값은 변경 가능          |
| 접근 방식   | 인덱스로 접근        | 인덱스로 접근        | 키로 접근                  |
| 사용 예     | 데이터 모음          | 고정된 항목 묶음      | 개체의 속성 표현           |


```python
# 3가지 자료구조 사용 예시
menu = ["coffee", "tea", "juice"]
screen_size = (1920, 1080)
user = {
    "username": "admin",
    "password": "1234",
    "active": True
}

print(menu[1])
print(screen_size[0])
print(user["username"])
```
